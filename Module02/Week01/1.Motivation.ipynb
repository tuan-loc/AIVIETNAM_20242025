### Example 1: Simple operations

```python
def f_function(number):
    result = number**2 + 1
    return result

def f_function_list(numbers):
    result = []
    for n in numbers:
        result.append(f_function(n))
    return result

# test
numbers = [1, 2, 3, 4]
print(f_function_list(numbers))
```

```python
import numpy as np

def f_function(number):
    result = number**2 + 1
    return result

factorial_np = np.vectorize(f_function)

# test
numbers = [1, 2, 3, 4]
numbers_array = np.array(numbers)
print(factorial_np(numbers_array))
```

### Compute factorial for a list

```python
def factorial(number):
    result = 1
    for i in range(2, number + 1):
        result *= i
    return result

def factorial_list(numbers):
    result = []
    for n in numbers:
        result.append(factorial(n))
    return result

# test
numbers = [1, 2, 3, 4]
print(factorial_list(numbers))
```

```python
import numpy as np

def numpy_factorial(numbers):
    # Create an array of ones with the same shape as the input
    result = np.ones_like(numbers, dtype=np.int64)
    
    # Iterate over each number in the input array
    for i in range(2, np.max(numbers) + 1):
        # Multiply result by i where numbers >= i
        result *= np.where(numbers >= i, i, 1)
    
    return result

# test
numbers = [1, 2, 3, 4]
numbers_array = np.array(numbers)
print(numpy_factorial(numbers_array))
```
