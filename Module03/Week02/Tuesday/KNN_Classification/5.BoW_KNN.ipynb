```python
import sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
```

```python
vectorizer = CountVectorizer()
corpus = ["góp gió gặt bão",
          "có làm mới có ăn",
          "đất lành chim đậu",
          "ăn cháo đá bát",
          "gậy ông đập lưng ông",
          "qua cầu rút ván"]
X = vectorizer.fit_transform(corpus)
```

```python
X = X.toarray()
print(X)
```

```python
x_test = vectorizer.transform(['không làm cạp đất mà ăn']).toarray()
x_test
```

```python
y_data = np.array([1, 1, 1, 0, 0, 0])
y_data
```

### Step-by-step

```python
# 1. compute distances
distances = np.sqrt( ((X-x_test)**2).sum(axis=1) )
print(distances)
```

```python
# 2. get the top k
top_k = np.argsort(distances)
print(top_k)

k = 3
top_k = top_k[:k]
print(top_k)
```

## sklearn

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=6)
classifier.fit(X, y_data)

y_pred = classifier.kneighbors(x_test)
print(y_pred)
```
