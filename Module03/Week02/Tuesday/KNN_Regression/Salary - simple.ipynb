```python
import numpy as np
import pandas as pd

df = pd.read_csv('Salary_Data_simple.csv')
df
```

```python
X = df.iloc[:, :-1].to_numpy()
y = df.iloc[:, -1:].to_numpy()
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X, y)

x_test = 4.3
predicted_salary = knn.predict([[x_test]])
print(predicted_salary)
```

```python
import numpy as np

# Define the test data point
x_test = np.array([4.3])

# Step 1: Calculate distances between x_test and all points in X
distances = np.sqrt(np.sum((X - x_test)**2, axis=1))

# Step 2: Find indices of k nearest neighbors
k = 3
nearest_neighbor_indices = np.argsort(distances)[:k]

# Step 3: Get the corresponding y values
nearest_neighbor_salaries = y[nearest_neighbor_indices]

# Step 4: Calculate the average (mean) of nearest neighbors' salaries
predicted_salary = np.mean(nearest_neighbor_salaries)
print(predicted_salary)
```
